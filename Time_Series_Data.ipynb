## Time resampling

#### https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78/ -->Read this for more info

```python
df.head()
```

```python
## Year end frequency

df.resample(rule='A').min()
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').max()['Open'].plot()
```

```python
## For quaterly start frequency

df.resample(rule='QS').max()
```

```python
df.resample(rule='QS').max()['High'].plot()
```

```python
## Bussiness End Frequency 

df.resample(rule='BA').max()['High'].plot()
```

```python
df.resample(rule='BQS').max()
```

```python
## Morden Syntax

##df.resample('D').max()
```

```python
## Plotting

df['Open'].resample('BA').mean().plot(kind='bar')
```

```python
df['Open'].resample('M').mean()
```

```python
df['Open'].resample('M').max().plot(kind='bar',figsize=(15,6))
```

### Rolling

```python
df['High'].rolling(10).mean().head(20)
```

```python
df['Open:30 days rolling']=df['Open'].rolling(30).mean()
```

```python
df.head(50)
```

```python
df[['Open','Open:30 days rolling']].plot(figsize=(12,5))
```

```python
## Rolling helps to smoothing

```

```python
df['ExpandingMean'] = df['Close'].expanding().mean()
```

```python
df[['Close','ExpandingMean']].plot(figsize=(12,5))
plt.title("Expanding Mean vs Original Close")
plt.show()
```
